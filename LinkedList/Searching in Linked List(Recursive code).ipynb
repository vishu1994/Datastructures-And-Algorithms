```python
class Node:
    def __init__(self,data):
        self.data=data
        self.next=None

class LinkedList :
    nth_index=0
    
    def __init__(self):
        self.head=None
    
    #insertion at beginning of the list....   
    def push(self,data):
        newNode=Node(data)
        newNode.next=self.head
        self.head=newNode
        
    #insertion at the end of the list.....
    def append(self,data):
        
        newNode=Node(data)
        if self.head is None :
            self.head=newNode
            return
        
        last=self.head
        
        while last.next :
            last=last.next
            
        last.next=newNode
    
    #insertion after some given node...
    def insertAfter(self,prev_node,data):
        
        newNode=Node(data)
        if prev_node is None :
            print("The given node doesnt exist")
            return
        
        temp=self.head
        
        while temp.data is not prev_node :
            temp=temp.next
        
        newNode.next=temp.next
        temp.next=newNode
    
    def printList(self):
        temp=self.head
        
        while temp:
            print(temp.data,end=" ")
            temp=temp.next
            
    def search(self,node,key):
        if node is None:
            print("search unsuccessful")
        
        elif node.data is key:
            print("search successful")
        
        else :
            return self.search(node.next,key)
    
    
    def searchRec(self,key):
        return self.search(self.head,key)

        

        
if __name__ =='__main__':
    
    llist=LinkedList()
    llist.append(1)
    llist.append(2)
    llist.append(3)
    llist.append(4)
    llist.append(5)
    llist.append(6)
    llist.append(7)
    
    
    llist.printList()
    print()
    
    llist.searchRec(4)
    print()

    llist.searchRec(0)
    
```
