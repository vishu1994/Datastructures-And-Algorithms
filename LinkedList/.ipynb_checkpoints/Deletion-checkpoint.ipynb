```python
class Node:
    def __init__(self,data):
        self.data=data
        self.next=None

class LinkedList :
    nth_index=0
    
    def __init__(self):
        self.head=None
    
    #insertion at beginning of the list....   
    def push(self,data):
        newNode=Node(data)
        newNode.next=self.head
        self.head=newNode
        
    #insertion at the end of the list.....
    def append(self,data):
        
        newNode=Node(data)
        if self.head is None :
            self.head=newNode
            return
        
        last=self.head
        
        while last.next :
            last=last.next
            
        last.next=newNode
    
    #insertion after some given node...
    def insertAfter(self,prev_node,data):
        
        newNode=Node(data)
        if prev_node is None :
            print("The given node doesnt exist")
            return
        
        temp=self.head
        
        while temp.data is not prev_node :
            temp=temp.next
        
        newNode.next=temp.next
        temp.next=newNode
    
    def printList(self):
        temp=self.head
        
        while temp:
            print(temp.data,end=" ")
            temp=temp.next
            
    
    def deleteAtBeginning(self):
        temp=self.head
        self.head=temp.next
        
        temp=None
    
    def delete(self,key):
        temp=self.head
        
        #if list is empty simply return
        if temp is None :
            return
        
        #if the head itself contains the key
        if temp.data == key :
            self.head=temp.next
            temp=None
            return
        
        while temp is not None:
            if temp.data== key :
                break;
            
            prev=temp
            temp=temp.next
        
        if temp is None:
            return
    
        prev.next=temp.next
        temp=None
            
        
    def deleteAt(self,pos):
        if self.head is None:
            return
        
        temp=self.head
        
        if pos is 0:
            self.head=temp.next
            temp=None
            return
        
        for i in range(pos-1):
            temp=temp.next
            if temp is None:
                break
        
        if temp is None:
            return
        
        temp.next=temp.next.next

        

        
if __name__ =='__main__':
    
    llist=LinkedList()
    llist.append(1)
    llist.append(2)
    llist.append(3)
    llist.append(4)
    llist.append(5)
    llist.append(6)
    llist.append(7)
    
    
    llist.printList()
    print()
   
    #Deletion at beginning of list....
    llist.deleteAtBeginning()
    llist.printList()
    print()
    
    llist.delete(7)
    llist.printList()
    
    
    
    
    
```
