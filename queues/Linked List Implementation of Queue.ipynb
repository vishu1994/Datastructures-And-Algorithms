```python
class QueueNode:
    
    def __init__(self,data):
        self.data=data
        self.next=None



class Queue:
    
    def __init__(self):
        self.front=None
        self.rear=None
    
    
    def isEmpty(self):
        
        return self.front==None
    
    def enQueue(self,data):
        
        newnode=QueueNode(data)
        
        if self.isEmpty():
            self.front=self.rear=newnode
            print("{} is inserted into the Queue ".format(data))
            return
        
        self.rear.next=newnode
        self.rear=newnode
        print("{} is inserted into the Queue ".format(data))
        
        
    
    def deQueue(self):
        
        if self.isEmpty():
            print("Queue Underflow ")
            return
        
        temp=self.front.data
        self.front=self.front.next
        print("{} is removed from the Queue ".format(temp))
         
    
    def queue_front(self):
        
        print("front of the Queue is :",self.front.data)
    
    def queue_rear(self):
        print("Rear of the Queue is :",self.rear.data)
    
    
    def printQueue(self):
        
        temp=self.front
        
        while temp:
            print(self.temp.data,end=" ")
            temp=temp.next
            

            
q=Queue()
q.enQueue(1)
q.enQueue(2)
q.enQueue(3)

q.queue_front()
q.queue_rear()

q.deQueue()

q.queue_front()
        
        
        
```
