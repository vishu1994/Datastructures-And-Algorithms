```python
class Queue:
    
    def __init__(self,capacity):
        self.size=self.front=0
        self.capacity=capacity
        self.rear=self.capacity-1
        self.q=[None]*self.capacity
        
    
    def isFull(self):
        
        return self.size==self.capacity
    
    def isEmpty(self):
        
        return self.size==0
    
    def enQueue(self,item):
        
        if self.isFull():
            print("Queue is Full ")
            return
        
        self.rear=(self.rear+1)%self.capacity
        self.q[self.rear]=item
        self.size=self.size+1
        print(" {} is inserted into the Queue ".format(item))
        
    
    def deQueue(self):
        
        if self.isEmpty():
            print("Queue undeflow ")
            return
        
        temp=self.q[self.front]
        self.front=(self.front+1)%self.capacity
        self.size=self.size-1
        
        print("{} is removed from the Queue".format(temp))
    
    
    def queue_front(self):
        
        print(" front element of the Queue is : ",self.q[self.front])
        
    
    def queue_rear(self):
        
        print(" rear element of the Queue is : ",self.q[self.rear])
        

q=Queue(5)
q.enQueue(1)
q.enQueue(2)
q.enQueue(3)
q.enQueue(4)
q.enQueue(5)

#q.enQueue(6)

q.queue_front()
q.queue_rear()


q.deQueue()
q.deQueue()

q.queue_front()
q.queue_rear()
    
    
```
