```python

dx=[-1,-1,-1,0,0,1,1,1]
dy=[-1,0,1,-1,1,-1,0,1]

def dfs(p,q,ar,n,m):
    #print("in dfs")
    if p>n-1 or p<0 or q>m-1 or q<0 or ar[p][q]==0:
        return
    
    ar[p][q]=0
    
    for a in range(8):
        dfs(p+dx[a],q+dy[a],ar,n,m)

ar=[]
t=int(input())
for i in range(t):
    
    rc=list(map(int,input().split()))
    n=rc[0]
    m=rc[1]
    for j in range(n):
        mystr=input()
        mylist=[]
        for ch in mystr:
            mylist.append(int(ch))
        ar.append(mylist)
    
    #This counts no of islands
    #print(ar)
    cnt=0
    for k in range(n):
        for l in range(m):
            #print(ar[k][l])
            print(k,l)
            if ar[k][l]==1:
                cnt+=1
                dfs(k,l,ar,n,m)
                
    print(cnt)
        
    #print(ar)
        
  
```
